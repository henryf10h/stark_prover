```python
import matplotlib.pyplot as plt
import numpy as np

# data points
data_points = [(0, 1), (1, 2), (2, 4), (3, 8)]
x_values, y_values = zip(*data_points)

# Lagrange polynomial
L = lambda x: -x**3/3 + 2*x**2 - 2*x + 8

# plot data points and polynomial
plt.scatter(x_values, y_values, color='r', label='Data Points')
x = np.linspace(min(x_values)-1, max(x_values)+1, 25)
plt.plot(x, L(x), color='b', label='Lagrange Polynomial')
plt.legend()
plt.show()
```
